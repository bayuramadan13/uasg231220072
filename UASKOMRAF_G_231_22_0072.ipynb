**A.) Distribusi data pada setiap atribut Purchase Price, Current Price dan Gain/Loss dengan grafik histrogram**

```python
# a.) Purchase Price
plt.figure(figsize=(15, 5))

# Mengambil data Purchase Price
plt.hist(dset['Purchase Price($)'], bins=13,
         color='cyan', edgecolor='white', alpha=0.9)

# Memberikan Judul, Mengubah Ukuran Font, dan Warna
plt.title('Histogram Untuk Purchase Price', loc='center',
          pad=20, fontsize=20, color='blue')
plt.xlabel('Purchase Price ($)', fontsize=15)
plt.ylabel('Frekuensi', fontsize=15, color='red')

# Memberikan angka terbatas di sumbu -y
plt.yticks(range(1, 11))

plt.show()
```

```python
# b.) Current Price

plt.figure(figsize=(15, 5))

# Mengambil data Current Price
plt.hist(dset['Current Price($)'], bins=13,
         color='cyan', edgecolor='white', alpha=0.9)

# Memberikan Judul, Mengubah Ukuran Font, dan Warna
plt.title('Histogram Untuk Current Price', loc='center',
          pad=20, fontsize=20, color='blue')
plt.xlabel('Current Price ($)', fontsize=15)
plt.ylabel('Frekuensi', fontsize=15, color='red')

# Memberikan angka terbatas di sumbu -y
plt.yticks(range(1, 11))

plt.show()
```

```python
# c.) Gain/Loss

plt.figure(figsize=(15, 5))

# Mengambil data Gain/Loss
plt.hist(dset['Gain/Loss($)'], bins=13,
         color='cyan' ,alpha=0.9)

# Memberikan Judul, Mengubah Ukuran Font, dan Warna
plt.title('Histogram Untuk Gain/Loss', loc='center',
          pad=20, fontsize=20, color='blue')
plt.xlabel('Gain/Loss ($)', fontsize=15)
plt.ylabel('Frekuensi', fontsize=15, color='red')

# Memberikan angka terbatas di sumbu -y
plt.yticks(range(1, 11))

plt.show()
```

**B. Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertinggi**

```python
# Mengambil data
sorted_dset = dset.sort_values('Dividend/share($)',
                             ascending=False)

# Memfilter data, dan memberi nomor urut
top5DS = sorted_dset.head(5)
top5DS = top5DS.set_index(pd.Index(range(1, len(top5DS)+1)))

# Memberi nama
hasil = top5DS[['Stock Name', 'Symbol','Dividend/share($)']]

print(hasil)
```

**C. Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi.**

```python
# Mengambil data
sorted_dset = dset.sort_values('Market Value($)',
                               ascending=False).reset_index(drop=True)

# Memfilter data, dan memberi nomor urut
top5MV = sorted_dset.head(5)
top5MV = top5MV.set_index(pd.Index(range(1, len(top5MV)+1)))

# Memberi nama
hasil = top5MV[['Stock Name', 'Symbol','Market Value($)']]

print(hasil)
```

**D. Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertinggi.**

```python
# Mengambil data
sorted_dset = dset.sort_values('Annual Yield',
                               ascending=False)

# Memfilter data, dan memberi nomor urut
top7AY = sorted_dset.head(7)
top7AY = top7AY.set_index(pd.Index(range(1, len(top7AY)+1)))

# Memberi nama
hasil = top7AY[['Stock Name', 'Symbol','Annual Yield']]

print(hasil)
```
